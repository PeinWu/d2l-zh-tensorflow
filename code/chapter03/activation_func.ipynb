```python
import tensorflow as tf
import matplotlib.pyplot as plt
tf.__version__
```

#### 生成数据

```python
x = tf.range(-10,10,0.1,dtype=tf.float32)
```

```python
def softmax(X):
    x_exp = tf.math.exp(X)
    sum_exp = tf.reduce_sum(x_exp)
    return x_exp / sum_exp
```

```python
y = softmax(x)
```

```python
plt.plot(x.numpy(),y.numpy())
```

```python
def sigmod(X):
    x_exp = tf.math.exp(-X)
    return 1/(1 + x_exp)
```

```python
x = tf.range(-20,20,dtype=tf.float32)
```

```python
y_2 = sigmod(x)
```

```python
plt.plot(x.numpy(),y_2.numpy())
```

```python
def tanh(X):
    x_exp_1 = tf.math.exp(X)
    x_exp_2 = tf.math.exp(-X)
    return (x_exp_1 - x_exp_2)/(x_exp_1+x_exp_2)
```

```python
y3 = tanh(x)
```

```python
plt.plot(x.numpy(),y3.numpy())
```
