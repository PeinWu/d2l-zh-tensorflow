```python
import tensorflow as tf
tf.__version__
```

### 构造模型

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(256,activation="relu"),
                            tf.keras.layers.Dense(10)])
```

```python
x = tf.random.uniform(shape=(2,20))
y = model(x)
```

```python
model.summary()
```

#### 访问模型

```python
model.layers[0].weights
```

```python
model.get_weights()
```

#### 参数初始化

```python
model2 = tf.keras.Sequential([tf.keras.layers.Dense(256,activation="relu",kernel_initializer=tf.keras.initializers.he_normal(),bias_initializer = tf.keras.initializers.zeros()),
                            tf.keras.layers.Dense(10,kernel_initializer = tf.keras.initializers.lecun_normal())])
```

```python
model2(x)
```

#### 自定义初始化

```python
def custon_init(shape,dtype):
    return tf.random.normal(shape,dtype=dtype)
```

```python
model3 = tf.keras.layers.Dense(20,kernel_initializer=custon_init,bias_initializer=custon_init)
```

```python
x = tf.random.normal(shape=(4,5))
model3(x)
```

```python
model3.weights
```

#### 参数共享
keras中参数共享实现方式是重复使用层

```python
dense = tf.keras.layers.Dense(32)

model = tf.keras.Sequential([
    dense,
    tf.keras.layers.Dense(16),
    dense,
    tf.keras.layers.Dense(2)
])
```

```python
left_x = tf.random.normal(shape=(4,3))

model(left_x)
```

```python
model.summary()
```
