### 自定义padding

```python
with tf.device("/cpu:0"):
    x = tf.random.normal(shape=(28,28,1))
    x = tf.expand_dims(x,axis=0)
    net = keras.layers.ZeroPadding2D(padding = (5,5))
    x = net(x)
    print(x.shape)
    net  = keras.layers.Conv2D(2, (3, 3))
    print(x.shape)
    y = net(x)
```

```python
y.shape
```

```python
with tf.device("/cpu:0"):
    x = tf.random.normal(shape=(28,28,1))
    x = tf.expand_dims(x,axis=0)
    net = keras.layers.ZeroPadding2D(padding = (5,5))
    x = net(x)
    print(x.shape)
    net  = keras.layers.Conv2D(2, (3, 3),strides=(2,1))
    print(x.shape)
    y = net(x)
```

```python
y.shape
```
